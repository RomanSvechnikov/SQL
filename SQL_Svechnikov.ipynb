# Задание 1
Необходимо с помощью SQL запроса собрать одну общую таблицу из всех данных витрин, которая должна включать в себя все поля этих витрин.
В качестве ответа предоставить ссылку на вашу копию страницы, где можно увидеть написанный запрос и собранную таблицу. Ссылку вставьте в Word файл "Техническое задание" в соответствующее поле.


_(одна из таблиц называется `group`, её имя конфликтует с зарезервированными ключевым словом SQL, поэтому его нужно будет заключить в двойные кавычки `""`)_


## Решение

Вывод всех витрин

```python
%%sql

SELECT mngmnt.org_management_rk, mngmnt.management_nm,
       "group".org_group_rk, "group".org_management_rk, "group".group_nm,
       emp_x_org_gr.employee_rk, emp_x_org_gr.org_group_rk,
       horoscope.agent_login, horoscope.horoscope,
       queue.queue_id, queue.queue_desc,
       task.task_rk, task.task_stage_id, task.source_system_cd, task.create_dttm, task.finish_dttm,
       action.hit_rk, action.hit_status_result_id,
       result.hit_status_result_id, result.hit_status_result_desc,
       product.hit_rk, product.hid, product.using_flg,
       call.wo_hit_rk, call.wo_task_rk, call.wo_queue_id, call.agent_login, call.wo_employee_rk, call.finish_dttm, call.duratoin_sec
FROM call JOIN mngmnt ON "group".org_management_rk = mngmnt.org_management_rk
          JOIN "group" ON emp_x_org_gr.org_group_rk = "group".org_group_rk
          JOIN emp_x_org_gr ON call.wo_employee_rk = emp_x_org_gr.employee_rk
          JOIN horoscope ON call.agent_login = horoscope.agent_login
          JOIN product ON call.wo_hit_rk = product.hit_rk
          JOIN action ON call.wo_hit_rk = action.hit_rk
          JOIN result ON action.hit_status_result_id = result.hit_status_result_id
          JOIN task ON call.wo_task_rk = task.task_rk
          JOIN queue ON call.wo_queue_id = queue.queue_id

LIMIT 10
```

# Задание 2
Необходимо посчитать следующие значения:

* Общее количество звонков со статусом «Дозвон, Успешно»;
* Продажная конверсия, по формуле:

  $$ Конверсия = \frac{N_{Дозвон, Успешно}}{N_{Дозвон, Успешно} + N_{Дозвон, Отказ}}$$
* Средняя длительность одной коммуникации, закрытой статусом «Дозвон, Успешно»

Ответы необходимо вписать в соответствующую таблицу в фаиле "Техническое задание"


## Решение

Верное решение

```python
%%sql
SELECT count, ROUND(CAST((count)*100 as REAL)/(count+N2), 2) as K, AVG
FROM
       (SELECT
       COUNT(result.hit_status_result_desc) FILTER (WHERE result.hit_status_result_desc = "Дозвон, Успешно") as "count",
       ROUND(AVG(call.duratoin_sec) FILTER (WHERE result.hit_status_result_desc = "Дозвон, Успешно"), 2) as "AVG",
       COUNT(result.hit_status_result_desc) FILTER (WHERE result.hit_status_result_desc = "Дозвон, Отказ") AS N2
       FROM result JOIN action ON action.hit_status_result_id = result.hit_status_result_id
                   JOIN call ON call.wo_hit_rk = action.hit_rk
       )
```

Средняя длительность одной коммуникации

```python
%%sql

SELECT round(AVG(call.duratoin_sec), 2) as "AVG"
FROM result JOIN action ON action.hit_status_result_id = result.hit_status_result_id
            JOIN call ON call.wo_hit_rk = action.hit_rk
WHERE hit_status_result_desc="Дозвон, Успешно"
```

Общее количество звонков со статусом «Дозвон, Успешно»

```python
%%sql

SELECT
COUNT(result.hit_status_result_desc) FILTER (WHERE result.hit_status_result_desc = "Дозвон, Успешно") as "count"
FROM result JOIN action ON action.hit_status_result_id = result.hit_status_result_id
```
